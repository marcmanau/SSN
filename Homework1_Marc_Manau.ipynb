Tornem a definir les diferents matrius que farem servir, utilitzarem diferents lletres per no portar a confusió, i calcularem a partir de la comanda .inv() l'inversa de la matriu.

```python
#Definim la matriu
E=sp.Matrix([[2,2,3],[1,-1,0],[-1,2,1]])
E[:3,:3]
```

```python
#Calculem l'inversa
E.inv()
```

Ara realitzarem l'equació de matrius per trobar el valor de la matriu incógnita X.

```python
#Definim la matriu de l'equació.
F=sp.Matrix([[1,2],[0,1],[5,4]])
F[:3,:2]
```

```python
#Elevem al quadrat la matriu principal.
E2=E**2
E2[:3,:3]
```

```python
#Fem la matriu inversa per poder obtenir la matriu X.
IE2=E2.inv()
X=IE2*F
X[:3,:2]
```

A continuació calcularem la incògnita de l'altra equació fent servir també l'altra matriu.

```python
#Definim la matriu.
G= sp.Matrix([[-1,1/3],[0,1/3]])
G[:2,:2]
```

```python
#Fem l'inversa i definim la matriu resultant de l'equació.
IG=G.inv()
H=Matrix([[-1,1],[0,1/3],[-5,5/3]])
H[:3,:2]
```

```python
#Resolem l'equació i obtenim la matriu Y.
Y=IE2*H*IG
Y[:3,:2]
```

## Exercici 3

a) Calculeu el polinomi característic de  M .
b) Determineu els seus valors propis i les multiplicitats corresponents.
c) Determineu els espais de vectors propis associats a cada un dels valors propis.
d) Comproveu que la matriu  M  és diagonalitzable.

```python
#Primer de tot, definim la matriu M amb la que treballarem a l'exercici.
M=sp.Matrix([[0.5,-0.5,-0.5,-0.5,1],[3.5,4.5,1.5,3.5,-1.5],[-0.5,-2.5,0.5,-0.5,1.5],[-5,-4,0,-6,6],[0,1,1,-2,3]])
M[:5,:5]
```

Per calcular el polinomi característic restarem x en la diagonal principal de la matriu i després farem el determinant.

```python
#Restem x en la diagonal principal.
x = sp.Symbol('x')
PM = M - x*eye(5)
PM[:5,:5]
```

```python
#Treiem el polinomi fent el determinant.
PM.det()
```

Per obtenir els valors propis farem servir la comanda .eigenvals(), al costat de cada valor propi apareixerà la seva multiplicitat.

```python
M.eigenvals()
```

I per calcular els vectors propis associats a cada valor propi farem servir M.eigenvects().

```python
M.eigenvects()
```

Finalment, diagonalitzarem la matriu fent servir .diagonalize(), obtindrem dues matrius diferents, una que anomenarem P que conté tots els vectors propis i una altra que serà DM que correspondrà a una matriu amb tots els valors propis en la diagonal. Amb aquestes dues matrius es dona la diagonalització.

```python
P, DM = M.diagonalize()
P[:,:]
```
